Generation of Synthetic Data
============================

```python
# We will use these for plotting the truth; these are not samples for our model.
xs = np.arange(0, 1, 0.01)
ys = np.sin(12 * (xs + 0.2)) / (xs + 0.2)

# This is the first sample containing 10 points.
X1 = np.random.random(10)
Y1 = np.sin(12 * (X1 + 0.2)) / (X1 + 0.2) + np.random.standard_normal(10)

# This is the second sample containing 100 points.
X2 = np.random.random(100)
Y2 = np.sin(12 * (X2 + 0.2)) / (X2 + 0.2) + np.random.standard_normal(100)

# Let's have a look at the data.
pylab.plot(xs, ys, 'k-', label='truth')
pylab.plot(X1, Y1, 'ro', label='small set')
pylab.plot(X2, Y2, 'ko', label='bigger set', alpha=.2)
pylab.legend()
```

Pick and normalize a data set
=============================

```python
# Let us pick one of the data sets first.
#X, Y = X1, Y1[:, np.newaxis]      # For the small data set.
X, Y = X2, Y2[:, np.newaxis]       # For the big data set.

# Normalize data by centering and setting std to 1.
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X - X_mean) / X_std
xs_ = (xs - X_mean) / X_std    # Yes, we are copying the xs here to avoid repetitive normalization each time we execute this panel.

Y_mean = Y.mean(axis=0)
Y_std = Y.std(axis=0)
Y = (Y - Y_mean) / Y_std
ys_ = (ys - Y_mean) / Y_std    # Yes, we are copying the ys here to avoid repetitive normalization each time we execute this panel.
```
